```python
import os
```

```python
masks=os.listdir(os.path.join('train','satellite'))
```

```python
masks
```

```python
lis=[]
for m in masks:
    if m[-3:]=='png':
        lis.append(m)
```

```python
len(lis)
```

```python
import cv2
import numpy as np
```

```python
co=0
for l in lis:
    mask=cv2.imread(os.path.join('train','satellite',l))
    #print(type(mask))
    #print(mask)
    co+=1
    m=np.zeros((1948,2048))
    print(co)
    for i in range(2048):
        for j in range(1948):
            if mask[j,i,0]==252 and mask[j,i,1]==0 and mask[j,i,2]==255:
                m[j,i]=255
    cv2.imwrite(os.path.join('updatedmasks',l),m)
    #print(mask)
    #cv2.imshow(mask)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
```
